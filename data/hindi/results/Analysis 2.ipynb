# Analysis 2
I've cleaned the data from McGregor a bit--got rid of prefixes and suffixes, replaced the weakened schwa /áµŠ/ with the regular schwa, and got rid of dashes indicating morpheme boundaries. Now I ask: how volatile are the classifications produced by the phonotactic RNNs?

```python
from numpy import argmin

def calc(data, j):
    pa = []
    sa = []
    na = []
    drav = []
    none = []
    en = []
    for i in data:
        if i not in translit: continue
        if translit[i] not in etym: continue
        if etym[translit[i]] == ['Perso-Arabic']: pa.append(data[i])
        if etym[translit[i]] == ['Sanskrit']: sa.append(data[i])
        if etym[translit[i]] == ['Native']: na.append(data[i])
        if etym[translit[i]] == ['Dravidian']: drav.append(data[i])
        if etym[translit[i]] == ['']: none.append(data[i])
        if etym[translit[i]] == ['English']: en.append(data[i])
    res = [pa, sa, na, drav, none, en]
    ans = []
    for i in range(j):
        ans.append([len([1 for x in y if argmin(x) == i]) / len(y) for y in res])
    return ans
```

```python
import csv
import matplotlib
import pandas as pd

translit = {}
with open('extra_large.csv', 'r') as translits:
    reader = csv.reader(translits)
    for row in reader:
        translit[row[1]] = row[0]
        
etym = {}
with open('mcgregor_etym.csv', 'r') as etyms:
    reader = csv.reader(etyms, delimiter='\t')
    for row in reader:
        etym[row[0]] = list(row[1].split(','))
```

```python
result = {}
with open('results2_2.txt', 'r') as results:
    reader = csv.reader(results, delimiter='\t')
    for row in reader:
        result[row[0]] = [float(x) for x in row[1:]]
```

```python
from pandas import DataFrame

a = DataFrame(calc(result, 2))
a.columns = ['Perso-Arabic', 'Sanskrit', 'Native', 'Dravidian', 'None', 'English']
a
```
